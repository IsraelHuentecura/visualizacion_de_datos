### <font color='#c38335'>Anatomia de una Figura</font>


```python
import matplotlib.pyplot as plt
img = plt.imread("../clases/images/anatomiafigura.png")
fig, ax = plt.subplots()
fig.set_size_inches(10,10)
plt.imshow(img)
```

<p style="font-size:14px">La <b>figura/figure</b> es el espacio/el contenedor dentro del cual vamos a crear una o más gráficas, cada una de ellas representada o accesible vía un conjunto de ejes. La figura contiene todos y cada uno de los atributos como los ejes, titulos, etc. Al crear una figura siempre existirá.</p>

Los <b>ejes/axes</b> contendran las gráficas y consiste en darle el entorno a dichas gráficas. Los ejes no son solo las líneas ortogonales que definen las referencias para las dimensiones invlucradas sino también tiene atributos como el título del gráfico, los ticks, las etiquetas, las leyendas.

Los <b>gráficos/plots</b> son los elementos que contiene los datos que queremos mostrar. El color del gráfico, el ancho de línea usada, etc., son atributos de los plots.

<b>Observación:</b> Para crear un gráfico es necesario crear una figura y, al menos, un conjunto de ejes.

```python
plt.figure(figsize=[9,6],facecolor="#fed8b1",edgecolor="#96603a",linewidth=10)
plt.suptitle('Anatomia de una figura',fontsize=16)
plt.axes()
plt.show()
```

La función <b>PLOT</b> crea de forma automática una figura y un conjunto de ejes. Esta forma es muy sencilla pero es la que menor capacidad de personalización que ofrece.

```python
plt.plot([2,6,7,3,9])
plt.show()
```

La función <b>FIGURE</b> y <b>AXES</b> crea de forma explícita una figura y un conunto de ejes en la última figura que se haya creado o referenciado.

```python
fig = plt.figure()
ax = plt.axes()
ax.plot([2,3,5,8,1])
plt.show()
```

La función <b>SUBPLOTS</b> crea una figura y uno o mas conjunto de ejes. Por defecto crea un solo axes.

```python
fig, ax = plt.subplots()
```

### <font color='#c38335'>Interfaces de programación</font>

Un tema que suele confundir al usuario de matplotlib es la variedad de métodos que hay para conseguir un objetivo. Y es que matplotlib ofrece dos interfaces de programación (digamos, dos estilos distintos para crear y personalizar las gráficas): estilo <b>MATLAB</b> y estilo <b>"Orientado a Objetos"</b>.

#### MATLAB

Continuidad

```python
plt.figure()

plt.subplot(1,2,1)
plt.plot([2,3,4,5])
plt.title("Left")
plt.xticks(range(0,4),["tick1","tick2","tick3","tick4"])

plt.subplot(1,2,2)
plt.scatter([1,2,3],[1,5,5])
plt.title("Right")

plt.show()
```

#### Orientado a Objetos
Vamos a referenciar la figura y los ejes mediante variables y serán métodos y atributos de estas variables los que nos permitirán personalizar la visualización.

```python
fix, ax = plt.subplots(1,2)

ax[0].plot([2,3,4,5])
ax[0].set_title("Left")
ax[0].set_xticks(range(0,4),["tick1","tick2","tick3","tick4"])

ax[1].scatter([1,2,3],[1,5,5])
ax[0].set_title("Right")

plt.show()
```

La mayor parte de los métodos que nos permiten fijar un valor a un atributo comienza por <b>"set_"</b>. Estos métodos suelen tener un método equivalente que comienza habitualmente por <b>"get_"</b> que sirve para leer el valor de dicho atributo.

Este segundo estilo orientado a objetos nos da mayor control sobre las visualizaciones que creemos.

### <font color='#c38335'>La función pcolormesh e imshow</font>

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pcolormesh.html

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.imshow.html

Diferencias:
* Las cabeceras de las funciones no son las mismas: aparte de otros parámetros, la función <b>imshow</b> acepta como primer parámetro un array (o estructura semejante) conteniendo los datos a mostrar. <b>pcolormesh</b> acepta las posiciones $x$ e $y$ de los bloques a colorear (ambos parámetros opcionales) y los datos en sí (uno menos por dimensión que $x$ e $y$).

matplotlib.pyplot.imshow(X, ...)

pcolor([x, y,] c, ...)

* Ambas funciones muestran los datos de izquierda a derecha, pero la función <b>imshow</b> considera el origen de coordenadas en la esquina superior izquierda -convención habitual en procesamiento de imágenes- mientras que <b>pcolormesh</b> lo considera en la esquina inferior izquierda.
* Los elementos coloreados son píxels en el caso de <b>imshow</b> y rectángulos en el caso de <b>pcolormesh</b>.
* Los elementos coloreados se muestran en posiciones $(i, j)$ -siendo $i$ y $j$ números enteros consecutivos a partir de $0$- en <b>imshow</b>, y, en <b>pcolormesh</b>, en las posiciones dadas por los arrays $x$ e $y$ con los que se llama a la función. Si no se incluyen los parámetros $x$ e $y$, entonces se muestran también en posiciones de $0$ en adelante.
* La función <b>imshow</b> está diseñada para mostrar imágenes, lo que supone que el array de entrada deberá tener tamaño $(m, n)$, $(m, n, 3)$ o $(m, n, 4)$. <b>pcolormesh</b> acepta solo arrays de dos dimensiones $(m, n)$.
