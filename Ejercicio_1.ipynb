```python
import matplotlib
from matplotlib import transforms, pyplot as plt
import numpy as np

# Definir colores
AZUL1, AZUL2, AZUL3, AZUL4 = '#1F77B4', '#AEC7E8', '#FF7F0E', '#FFBB78'
VERDE1, VERDE2 = '#2CA02C', '#98DF8A'
GRIS1, GRIS2, GRIS3, GRIS4 = '#D62728', '#FF9896', '#9467BD', '#C5B0D5'


# Aos
X = ['Hoy\n30/09/24', 'FY25', 'FY26', 'FY27', 'FY28', 'FY29']
COSTO = np.array([0, -22, -34, -62, -77, -99])  # Costo
INGRESO = np.array([190, 162, 146, 128, 113, 91])  # Ingreso
GANANCIA = np.array([0, 7, 10, 14, 16, 22])  # Ganancia
OTROS_GASTOS = np.array([0, 4, 4, 6, 8, 9])  # Otros gastos
PERDIDA = np.array([0, 28, 45, 56, 65,78])  # Prdida

# Crear nueva figura
fig, ax1 = plt.subplots(figsize=(10, 7))  # Ancho, alto en pulgadas

# Establecer los lmites de datos para los ejes y, x
ax1.set_ylim([-100, 250])
ax1.set_xlim([-0.5, 5.4])
# Mover los ejes x a cero
ax1.spines['bottom'].set_position('zero')

plt.bar(list(range(0, 6)), COSTO, width=0.68, color=AZUL4, edgecolor=AZUL4, linewidth=0.9)
plt.bar(list(range(0, 6)), INGRESO, width=0.68, color=AZUL2, edgecolor=AZUL2, linewidth=0.9)
plt.bar(list(range(0, 6)), GANANCIA, width=0.68, bottom=INGRESO, color=VERDE1, edgecolor=VERDE1, linewidth=0.9)
plt.bar(list(range(0, 6)), OTROS_GASTOS, width=0.68, bottom=INGRESO + GANANCIA, color=VERDE2, edgecolor=VERDE2, linewidth=0.9)
white_bars = plt.bar(list(range(0, 6)), PERDIDA, width=0.68, bottom=INGRESO + GANANCIA + OTROS_GASTOS, color='white', edgecolor=GRIS3, linewidth=0.9)

# Agregar etiquetas de datos numricos para las barras
for i, b in enumerate(white_bars):
    if PERDIDA[i] == 0:
        continue
    x = b.get_x() + b.get_width() / 2  # Centro x de la barra
    y = b.get_y() + b.get_height() - 20  # Parte superior y de la barra
    ax1.text(x, y, str(PERDIDA[i]), fontsize=20, fontweight='bold', horizontalalignment='center', color=GRIS1)

# Agregar texto para las etiquetas de datos numricos
ax1.text(5.5, y, 'Prdida', fontsize=20, fontweight='bold', color=GRIS1)

# Agregar texto etiquetando otras barras
ax1.text(5.5, 110, 'Gastos adicionales', fontsize=16, color=VERDE2)
ax1.text(5.5, 95, 'Ganancias', fontsize=16, color=VERDE1)
ax1.text(5.5, 80, 'Ingreso', fontsize=16, color=AZUL2)
ax1.text(5.5, -100, 'Costo', fontsize=16, color=AZUL4)

# Establecer propiedades para el objeto de ejes (marcas para todos los problemas con etiquetas)
plt.setp(ax1, xticks=list(range(0, 6)), xticklabels=X)

# Eliminar borde del grfico
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.spines['bottom'].set_color(GRIS3)
ax1.spines['left'].set_color(GRIS3)
ax1.spines['left'].set_linewidth(1.5)
ax1.spines['bottom'].set_linewidth(1.5)

# Cambiar color de las marcas y
ax1.tick_params(axis='y', colors=GRIS3)
# Ocultar marcas x
ax1.tick_params(bottom='off')

# Configurar apariencia de etiquetas de marcas x
for item in ax1.get_xticklabels():
    item.set_fontsize(14)
    item.set_color(AZUL1)

# Configurar apariencia de etiquetas de marcas y
for item in ax1.get_yticklabels():
    item.set_fontsize(14)
    item.set_color(GRIS1)

# Agregar ttulo y etiquetas
# Bajar un poco hacia abajo el label de ao
ax1.set_xlabel('Ao', fontsize=16, color=GRIS2)
ax1.set_ylabel('Dinero (en millones de pesos)', fontsize=16, color=GRIS2)
plt.title('Rentabilidad Esperada a lo Largo del Tiempo', fontsize=20, color=GRIS2, pad=20)

# Agregar footnote con la fuente de datos
plt.text(-0.75, -137, 'Fuente: Datos financieros de la empresa', fontsize=12, color=GRIS3)
# Agregar Caption con el autor pero fuera de la figura
plt.text(-0.75, -170, 'Figura 1: Rentabilidad esperada a lo largo del tiempo', fontsize=12, color='black')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Definir colores
AZUL1, AZUL2, AZUL3, AZUL4 = '#1F77B4', '#AEC7E8', '#FF7F0E', '#FFBB78'
VERDE1, VERDE2 = '#2CA02C', '#98DF8A'
GRIS1, GRIS2, GRIS3, GRIS4 = '#D62728', '#FF9896', '#9467BD', '#C5B0D5'

# Aos
X = ['Hoy\n30/09/24', 'FY25', 'FY26', 'FY27', 'FY28', 'FY29']
COSTO = np.array([0, -22, -34, -62, -77, -99])  # Costo
INGRESO = np.array([190, 162, 146, 128, 113, 91])  # Ingreso
GANANCIA = np.array([0, 7, 10, 14, 16, 22])  # Ganancia
OTROS_GASTOS = np.array([0, 4, 4, 6, 8, 9])  # Otros gastos
PERDIDA = np.array([0, 28, 45, 56, 65,78])  # Prdida

# Crear DataFrame
df = pd.DataFrame({'Ao': X*5,
                   'Valor': np.concatenate([COSTO, INGRESO, GANANCIA, OTROS_GASTOS, PERDIDA]),
                   'Tipo': np.repeat(['Costo', 'Ingreso', 'Ganancia', 'Otros gastos', 'Prdida'], len(X))})

# Crear nueva figura
plt.figure(figsize=(10, 7))

# Crear grfico de barras apiladas
sns.barplot(data=COSTO, color=AZUL4, edgecolor=AZUL4, linewidth=0.9)
sns.barplot(data=PERDIDA, color=GRIS1, edgecolor=GRIS1, linewidth=0.9, bottom=COSTO)
sns.barplot(data=INGRESO, color=AZUL2, edgecolor=AZUL2, linewidth=0.9)
sns.barplot(data=GANANCIA, color=VERDE1, edgecolor=VERDE1, linewidth=0.9)
sns.barplot(data=OTROS_GASTOS, color=VERDE2, edgecolor=VERDE2, linewidth=0.9)

# Establecer los lmites de datos para los ejes y, x
plt.ylim([-100, 250])
plt.xlim([-0.5, 5.4])


# Agregar una linea en el 0 horizontal
ax = plt.gca()
ax.axhline(y = 0, color = 'black', linewidth = 1.3, alpha = .7)

# Establecer que X son los nombres del eje x
plt.xticks(range(0, 6), X)

# Agregar ttulo y etiquetas
plt.xlabel('Ao', fontsize=16, color='#9b9b9b')
plt.ylabel('Dinero (en millones de pesos)', fontsize=16, color='#9b9b9b')
plt.title('Rentabilidad Esperada a lo Largo del Tiempo', fontsize=20, color='#9b9b9b', pad=20)

# Agregar leyenda con sus respectivos colores del grfico
# Agregar leyenda con sus respectivos colores del grfico
legend_labels = ['Prdida', 'Costo', 'Ingreso', 'Ganancia', 'Otros gastos']
legend_handles = [
    mpatches.Patch(color=GRIS1, label='Prdida'),
    mpatches.Patch(color=AZUL4, label='Costo'),
    mpatches.Patch(color=AZUL2, label='Ingreso'),
    mpatches.Patch(color=VERDE1, label='Ganancia'),
    mpatches.Patch(color=VERDE2, label='Otros gastos')
]
# Agregar leyenda
plt.legend(handles=legend_handles, labels=legend_labels)
# Agregar footnote con la fuente de datos
plt.text(-0.75, -170, 'Fuente: Datos financieros de la empresa', fontsize=12, color=GRIS3)
# Agregar Caption con el autor pero fuera de la figura
plt.text(-0.75, -220, 'Figura 1: Rentabilidad esperada a lo largo del tiempo', fontsize=12, color='black')

plt.show()
```
