```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Definir colores
AZUL1, AZUL2, AZUL3, AZUL4 = '#1F77B4', '#AEC7E8', '#FF7F0E', '#FFBB78'
VERDE1, VERDE2 = '#2CA02C', '#98DF8A'
GRIS1, GRIS2, GRIS3, GRIS4 = '#D62728', '#FF9896', '#9467BD', '#C5B0D5'

# Aos
X = ['Hoy\n30/09/24', 'FY25', 'FY26', 'FY27', 'FY28', 'FY29']
COSTO = np.array([0, -22, -34, -62, -77, -99])  # Costo
INGRESO = np.array([190, 162, 146, 128, 113, 91])  # Ingreso
GANANCIA = np.array([0, 7, 10, 14, 16, 22])  # Ganancia
OTROS_GASTOS = np.array([0, 4, 4, 6, 8, 9])  # Otros gastos
PERDIDA = np.array([0, 28, 45, 56, 65,78])  # Prdida

# Crear DataFrame
df = pd.DataFrame({'Ao': X*5,
                   'Valor': np.concatenate([COSTO, INGRESO, GANANCIA, OTROS_GASTOS, PERDIDA]),
                   'Tipo': np.repeat(['Costo', 'Ingreso', 'Ganancia', 'Otros gastos', 'Prdida'], len(X))})

# Crear nueva figura
plt.figure(figsize=(10, 7))

# Crear grfico de barras apiladas
sns.barplot(data=COSTO, color=AZUL4, edgecolor=AZUL4, linewidth=0.9)
sns.barplot(data=PERDIDA, color=GRIS1, edgecolor=GRIS1, linewidth=0.9, bottom=COSTO)
sns.barplot(data=INGRESO, color=AZUL2, edgecolor=AZUL2, linewidth=0.9)
sns.barplot(data=GANANCIA, color=VERDE1, edgecolor=VERDE1, linewidth=0.9)
sns.barplot(data=OTROS_GASTOS, color=VERDE2, edgecolor=VERDE2, linewidth=0.9)

# Establecer los lmites de datos para los ejes y, x
plt.ylim([-100, 250])
plt.xlim([-0.5, 5.4])


# Agregar una linea en el 0 horizontal
ax = plt.gca()
ax.axhline(y = 0, color = 'black', linewidth = 1.3, alpha = .7)

# Establecer que X son los nombres de los aos pero que salgan de manera horizontal en punto 0 de Y
plt.xticks(ticks=np.arange(6), labels=X, rotation=0, fontsize=12, color='#9b9b9b')
plt.yticks(fontsize=12, color='#9b9b9b')

# Eliminar eje derecho y superior
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)

# Agregar ttulo y etiquetas
plt.xlabel('Ao', fontsize=16, color='#9b9b9b')
plt.ylabel('Dinero (en millones de pesos)', fontsize=16, color='#9b9b9b')
plt.title('Rentabilidad Esperada a lo Largo del Tiempo', fontsize=20, color='#9b9b9b', pad=20)

# Agregar leyenda con sus respectivos colores del grfico
# Agregar leyenda con sus respectivos colores del grfico
legend_labels = ['Prdida', 'Costo', 'Ingreso', 'Ganancia', 'Otros gastos']
legend_handles = [
    mpatches.Patch(color=GRIS1, label='Prdida'),
    mpatches.Patch(color=AZUL4, label='Costo'),
    mpatches.Patch(color=AZUL2, label='Ingreso'),
    mpatches.Patch(color=VERDE1, label='Ganancia'),
    mpatches.Patch(color=VERDE2, label='Otros gastos')
]
# Agregar leyenda
plt.legend(handles=legend_handles, labels=legend_labels)
# Agregar footnote con la fuente de datos
plt.text(-0.75, -170, 'Fuente: Datos financieros de la empresa', fontsize=12, color='black')
# Agregar Caption con el autor pero fuera de la figura
plt.text(-0.75, -220, 'Figura 1: Rentabilidad esperada a lo largo del tiempo', fontsize=12, color='black')

plt.show()
```

```python


# Utilizar un grfico de seaborn para visualizar los datos
sns.set_theme(style='white')
# Quitar eje superior y derecho
sns.set(style='white', rc={'axes.spines.right': False, 'axes.spines.top': False})


# Cargamos el dataset
df = sns.load_dataset('titanic')

# Creamos los intervalos de edades
df['age_group'] = pd.cut(df['age'], bins=[0, 18, 26, 41, df['age'].max()], labels=['0-17', '18-25', '26-40', '41+'])

# Creamos una lista con los dataframes
dfs = [df[df['age_group'] == age_group] for age_group in ['0-17', '18-25', '26-40', '41+']]
titulos = ['0-17', '18-25', '26-40', '41+']

# Creamos una figura con 4 subplots en la columna izquierda y 1 subplot en la columna derecha
fig, axs = plt.subplots(4, 2, figsize=(20, 10), gridspec_kw={'width_ratios': [2, 2]})

# Hacemos un grfico para cada intervalo
for i, data in enumerate(dfs):
    # Grfico de densidad
    ax = axs[i, 0]
    for sex in ['male', 'female']:
        # Hacer ms largo de izquierda a derecha
        sns.kdeplot(data=data[data['sex'] == sex]['age'], label=sex, ax=ax, shade=True)
        # Quitar el nombre del eje y
        ax.set_ylabel('')
        
    # El titulo ponerlo en inferior a la izquierda
    
    ax.text(0.05, 0.3, f'Age: {titulos[i]}', transform=ax.transAxes, fontsize=16)
    ax.legend()

# Grfico de barras consolidado
ax = plt.subplot2grid((4, 2), (0, 1), rowspan=4)
# Barras horizontales
sns.countplot(data=df, y='age_group', hue='sex', ax=ax, palette='viridis')
# No mostrar nombre del eje Y
ax.set_ylabel('')

# Quitamos los ttulos de los ejes de la derecha
for i in range(4):
    axs[i, 1].set_xticks([])
    axs[i, 1].set_yticks([])
    axs[i, 1].set_xticklabels([])
    axs[i, 1].set_yticklabels([])
    axs[i, 1].set_xlabel('')
    
# Poner titulo general
fig.suptitle('Distribucin de Edades por Sexo', fontsize=20, color='#9b9b9b', y=1.02)

plt.tight_layout()
plt.show()
```
