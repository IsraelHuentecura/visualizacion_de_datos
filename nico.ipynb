```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Paso 1: Cargar y procesar los datos
data = {
    'datetime': [
        "19/3/2022 12:21", "19/3/2022 12:29", "19/3/2022 12:30", "19/3/2022 12:30", 
        # ... aadir todas las fechas y horas
    ],
    'sender': [
        "Bernardo", "Daniel Stampaz", "Daniel Stampaz", "Daniel Stampaz",
        # ... aadir todos los remitentes
    ],
    'message': [
        "Filtros", "Okas dokas...", "El motor vale 45 + IVA en stampaz", "El de RT que me preguntaste",
        # ... aadir todos los mensajes
    ]
}

df = pd.DataFrame(data)
df['datetime'] = pd.to_datetime(df['datetime'], format='%d/%m/%Y %H:%M')

# Paso 2: Visualizacin
# Lnea de tiempo de mensajes
plt.figure(figsize=(15, 10))
sns.lineplot(x='datetime', y='sender', data=df, marker='o')

plt.title('Lnea de tiempo de la conversacin')
plt.xlabel('Fecha y Hora')
plt.ylabel('Remitente')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Frecuencia de mensajes por remitente
plt.figure(figsize=(10, 6))
sns.countplot(y='sender', data=df, palette='viridis')

plt.title('Frecuencia de mensajes por remitente')
plt.xlabel('Cantidad de mensajes')
plt.ylabel('Remitente')
plt.show()
```
